```python
import nltk
import matplotlib.pyplot as plt
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
RomeoandJuliet = open('/content/drive/MyDrive/CO-OP Summer 2021/Text Analysis/RomeoandJuliet.txt').read().split("SCENE")
Macbeth = open('/content/drive/MyDrive/CO-OP Summer 2021/Text Analysis/Macbeth.txt').read().split("SCENE")
Othello = open('/content/drive/MyDrive/CO-OP Summer 2021/Text Analysis/Othello.txt').read().split("SCENE")
KingLear = open('/content/drive/MyDrive/CO-OP Summer 2021/Text Analysis/KingLear.txt').read().split("SCENE")
books = [RomeoandJuliet, Macbeth, Othello, KingLear]
titles = ['Romeo and Juliet', 'Macbeth', 'Othello', 'King Lear']
scenes = [[]]


sia.polarity_scores(RomeoandJuliet[0])
sia.polarity_scores(RomeoandJuliet[1])
polarityScores = []
for i in range(len(books)):
  polarityScores.append([])
  for j in range(len(books[i])):
    polarityScores[i].append(sia.polarity_scores(books[i][j]))


values = []
for i in range(len(polarityScores)):
  values.append([])
  for j in range(len(polarityScores[i])):
    values[i].append(polarityScores[i][j]['compound'])


for i in range(len(books)):
  plt.plot(range(len(values[i])), values[i])
  plt.title('Sentiment Value of Scenes in '+titles[i])
  plt.xlabel('Scene')
  plt.ylabel('Sentiment Value')
  plt.show()
```
